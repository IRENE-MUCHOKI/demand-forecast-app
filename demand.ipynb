```python
import os
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
```

```python
root_folder=r"C:\Users\irene\OneDrive\Desktop\python1"
```

```python
df=pd.read_csv(r"C:\Users\irene\OneDrive\Desktop\python1\demand_forecasting.csv")
df.head()
```

```python
df['Date'] = pd.to_datetime(df['Date'])
df=df.sort_values('Date')
df=pd.get_dummies(df, columns=['Store ID', 'Product ID', 'Category', 'Region', 'Weather Condition', 'Seasonality'])
df=df.drop("Date", axis=1)
df.head()
```

```python
df=df.drop("Epidemic", axis=1)
df.head()
```

```python
#data prep
y= df['Demand']
y
```

```python
x= df.drop('Demand', axis=1)
x
```

```python
#data split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
x_train
```

```python
x_test
```

```python
#model building
#linear regression
from sklearn.linear_model import LinearRegression
lr= LinearRegression()
lr.fit(x_train, y_train)
y_predr= lr.predict(x_train)
y_prede= lr.predict(x_test)
print("Predictions:", y_predr)
```

```python
print("Predictions:", y_prede)
```

```python
from sklearn.metrics import mean_absolute_error
preds=lr.predict(x_test)
mae= mean_absolute_error(y_test, preds)
print("Mean Absolute Error:", mae)
```

```python
#random forest
from sklearn.ensemble import RandomForestRegressor
rf= RandomForestRegressor(n_estimators=100, random_state=42)
rf.fit(x_train, y_train)
y_predrf= rf.predict(x_test)
print("Predictions:", y_predrf)
```
